```python
def lookup(coder, timeout, work, work_condition, lock, loc_dict, sig):
    name = None
    while name != 'SHUTDOWN':
        with work_condition:
            print(sig, end='')
            if len(work) > 0:
                name = work.pop()
            else:
                name = 'SHUTDOWN'
        if name not in loc_dict:
            timeout = False
            try:
                location = coder.geocode(name, timeout=timeout)
            except GeocoderTimedOut:
                location = None
                timeout = True
            if location is not None:
                lock.acquire()
                loc_dict[name] = location
                lock.release()
            elif not timeout and location is None:
                lock.acquire()
                loc_dict[name] = 'UNKNOWN'
                lock.release()

work_cond = Condition()
lookup_lock = Lock()
work = []
place_dict = dict()

for place in data["Place of death"]:
    if not pd.isnull(place):
        place_str = place + ', South Africa'
        work.append(place_str)

threads = []
for sig in ('.','*','-','^'):
    thread = Thread(target=lookup, args=(nom, 10, work, work_cond, lookup_lock, place_dict, sig))
    threads.append(thread)
    thread.start()

thread_count = len(threads)
for thread in threads:
    thread.join()

print('Done')
```
